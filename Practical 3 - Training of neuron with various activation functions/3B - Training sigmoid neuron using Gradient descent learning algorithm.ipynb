3B. Generate 51 points for = 1 1 + exp⁡(−3x),where x ∈ [−2, 3].Use this dataset to train
sigmoid neuron using gradient descent learning algorithm. Draw two curves with different
colours, for target and output(y) of the trained neuron.


```python
#3B
import numpy as np
import matplotlib.pyplot as plt

# Generating dataset
X = np.linspace(-2, 3, 51).reshape(-1, 1)
y = 1 + np.exp(-3 * X)

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Forward pass
def forward(X, w, b):
    return sigmoid(np.dot(X, w) + b)

# Training
lr = 0.1
epochs = 1000
w = np.random.randn(1, 1)
b = np.random.randn(1)
for epoch in range(epochs):
    y_pred = forward(X, w, b)
    error = y_pred - y
    w -= lr * np.dot(X.T, error)
    b -= lr * np.sum(error)

# Plotting
plt.plot(X, y, label='Target')
plt.plot(X, forward(X, w, b), label='Output')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sigmoid Neuron Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating dataset
x_values = np.linspace(-2, 3, 51)
y_values = 1 + np.exp(-3 * x_values)

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Weight and bias initialization
w = np.random.rand() 
b = np.random.rand()

# Learning rate
learning_rate = 0.01

# Number of epochs
epochs = 1000 

# Training loop
for epoch in range(epochs):
    # Compute the linear combination of inputs and weights
    z = w * x_values + b
    
    # Apply the sigmoid function
    predictions = sigmoid(z)
    
    # Calculate the loss (mean squared error for simplicity)
    loss = np.mean((predictions - y_values) ** 2)
    
    # Gradient of loss w.r.t. weights
    dW = np.dot((predictions - y_values) * predictions * (1 - predictions), x_values) / len(x_values)
    
    # Gradient of loss w.r.t. bias
    dB = np.mean((predictions - y_values) * predictions * (1 - predictions))
    
    # Update weights and bias
    w -= learning_rate * dW
    b -= learning_rate * dB

# Plotting
plt.plot(x_values, y_values, label='Actual Function')
plt.plot(x_values, sigmoid(w * x_values + b), label='Learned Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
