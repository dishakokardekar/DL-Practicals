```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1):
        self.weights = np.zeros(input_size)
        self.bias = 0
        self.learning_rate = learning_rate

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        return self.activation(weighted_sum)

    def train(self, training_inputs, labels, epochs):
        for _ in range(epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights += self.learning_rate * (label - prediction) * inputs
                self.bias += self.learning_rate * (label - prediction)

# AND Gate
def AND_gate(x1, x2):
    inputs = np.array([x1, x2])
    perceptron = Perceptron(2)
    perceptron.weights = np.array([0.5, 0.5])
    perceptron.bias = -0.7
    return perceptron.predict(inputs)

# OR Gate
def OR_gate(x1, x2):
    inputs = np.array([x1, x2])
    perceptron = Perceptron(2)
    perceptron.weights = np.array([0.5, 0.5])
    perceptron.bias = -0.2
    return perceptron.predict(inputs)

# NAND Gate
def NAND_gate(x1, x2):
    inputs = np.array([x1, x2])
    perceptron = Perceptron(2)
    perceptron.weights = np.array([-0.5, -0.5])
    perceptron.bias = 0.7
    return perceptron.predict(inputs)

# NOR Gate
def NOR_gate(x1, x2):
    inputs = np.array([x1, x2])
    perceptron = Perceptron(2)
    perceptron.weights = np.array([-0.5, -0.5])
    perceptron.bias = 0.2
    return perceptron.predict(inputs)

# XOR Gate
def XOR_gate(x1, x2):
    return AND_gate(NAND_gate(x1, x2), OR_gate(x1, x2))

# Testing NAND gate
print("NAND Gate:")
print("0 NAND 0 =", NAND_gate(0, 0))
print("0 NAND 1 =", NAND_gate(0, 1))
print("1 NAND 0 =", NAND_gate(1, 0))
print("1 NAND 1 =", NAND_gate(1, 1))

# Testing NOR gate
print("\nNOR Gate:")
print("0 NOR 0 =", NOR_gate(0, 0))
print("0 NOR 1 =", NOR_gate(0, 1))
print("1 NOR 0 =", NOR_gate(1, 0))
print("1 NOR 1 =", NOR_gate(1, 1))

# Testing XOR gate
print("\nXOR Gate:")
print("0 XOR 0 =", XOR_gate(0, 0))
print("0 XOR 1 =", XOR_gate(0, 1))
print("1 XOR 0 =", XOR_gate(1, 0))
print("1 XOR 1 =", XOR_gate(1, 1))
```
