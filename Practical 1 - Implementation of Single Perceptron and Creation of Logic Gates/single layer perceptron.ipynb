```python
#single layer perceptron
x_input=[0.1,0.5,0.2]
w_weight=[0.8,0.5,0.9]
threshold= 0.1
```

```python
def step (weighted_sum):
    if weighted_sum >= threshold:
        return 1
    else:
        return 0
```

```python
def perceptron():
    weighted_sum=0
    for x,w in zip(x_input,w_weight):
        weighted_sum += x*w
        print(weighted_sum)
    return step(weighted_sum)
```

```python
output = perceptron()
print("output=",output)
```

```python
# loss = n(d-y)x, where n=learning rate ,y=current output,d=expected outcome,x=inputs
#updated weight=w(new)=w1 + n(d-y)x
```
